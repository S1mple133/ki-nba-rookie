```python
nba_data.drop(['Brent Price', 'Jon Barry'], axis=0, inplace=True)
```

```python
print('Q1: ' + str(nba_data['GP'].quantile(0.25)))
print('Q3: ' + str(nba_data['GP'].quantile(0.75)))
print('Index mit dem höchsten GP Wert: ' + str(nba_data['GP'].idxmax()))
```

```python
nba_data['MIN'] = nba_data['MIN'] * 2
nba_data['MIN']
```

```python
def create_experience(row): 
  result = 'Unerfahren'
  if row['GP'] >= 70:
      if row['MIN'] >= 50.0: 
        result = 'Erfahren'
      else:
              result = 'Amateur'
  return result
nba_data['Spielerfahrung'] = nba_data.apply(create_experience, axis=1)
```
